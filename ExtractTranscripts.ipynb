```python
def ExtractText(id):
    with ZipFile("/Volumes/Macintosh HD - Data/Users/admin/Downloads/{}_P.zip".format(id), 'r') as zip:
        zip.extract("{}_TRANSCRIPT.csv".format(id), 'text')
        df = pd.read_csv('text/{}_TRANSCRIPT.csv'.format(id),sep = '\t')
        df['id'] = id
    print(f'adding participant {id} with shape {df.shape}')
    return df
```
